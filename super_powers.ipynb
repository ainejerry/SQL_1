```python
import pandas as pd 
```

```python
#loading the data into pandas 
df_1=pd.read_csv(r"C:\Users\DELL\Downloads\superhero_info - superhero_info.csv",low_memory=False)
df_1.head()
```

```python
#string split 
df_1[['hero_names', 'publisher']] = df_1['Hero|Publisher'].str.split('|', expand=True)
df_1.head()
```

```python
#dropping a column
df_1.drop('Hero|Publisher', axis=1, inplace=True)
df_1.head()
```

```python
#loading the 2nd data set
df_2=pd.read_csv(r"C:\Users\DELL\Downloads\superhero_powers - superhero_powers.csv",low_memory=False)
df_2.head()
```

```python
# Merge the dataframes based on the common column
df = pd.merge(df_1, df_2, on='hero_names')
df.head()
```

```python
#doing a split  based on the measurement column 
df[['Height', 'Weight']] = df['Measurements'].str.split(',', expand=True)
df.head()
```

```python
#column drop
df.drop('Measurements' ,axis=1,inplace=True)
df.head()
```

```python
#removing unnecessary characters in the height column
df['Height(cm)'] = df['Height'].str.extract(r'(\d+.\d+)', expand=False).astype(float)

df.head()
```

```python
#dropping a column
df.drop('Height' ,axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Weight(kg)'] = df['Weight'].str.extract(r'(\d+.\d+)', expand=False).astype(float)
df.head()
```

```python
#dropping a column
df.drop('Weight' ,axis=1,inplace=True)
df.head()
```

```python
#CHANGE OF COLUMN POSTION 
hm = df.pop('hero_names')  
df.insert(0, 'hero_names', hm) 
df.head()

```

```python
df['Powers'].value_counts()
```

```python
#encode the features in the powers column 
encode_cols=df['Powers'].str.get_dummies(',')
df=pd.concat([df,encode_cols],axis=1)
```

```python
df.head()
```

```python
#column drop 
df.drop('Powers' ,axis=1,inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting weight against publishers 
average_heights = df.groupby('publisher')['Height(cm)'].mean()

# Plotting the average heights
average_heights.plot(kind='bar')
plt.xlabel('Publisher')
plt.ylabel('Average Height')
plt.title('Average Height of Heroes by Publisher')
plt.xticks(rotation=45)
plt.show()
```

```python

#filter for superheroes with super speed
super_speed_heroes = df[df['Super Speed'] == True]

# Filter superheroes without Super Speed
non_super_speed_heroes = df[df['Super Speed'] == False]

# Calculate average weight for superheroes with Super Speed
average_weight_super_speed = super_speed_heroes['Weight(kg)'].mean()

# Calculate average weight for superheroes without Super Speed
average_weight_non_super_speed = non_super_speed_heroes['Weight(kg)'].mean()

# Plotting the average weights
plt.bar(['Super Speed', 'No Super Speed'], [average_weight_super_speed, average_weight_non_super_speed])
plt.xlabel('Super Speed')
plt.ylabel('Average Weight')
plt.title('Average Weight of Superheroes with and without Super Speed')
plt.show()
```
